3. Вивести середнє значення та відхилення курсу за кожний місяць.

```python
import datetime
import math
import requests

def gathering_info(year_: int, month_: int, day_: int, range_of_days: int) -> list:
    """Take start date and the range of days of the necessary period and return
    the list of currency rates from this period as floats.
    : params `year_, month_, day_`: start date of the necessary period input as 
    integers in format 'year, month, day'
    : param `range_of_days`: amount of days of the necessary period input as an integer
    """
    list_month_rate = []
    #     researched_period = year_, month_, day_
    start = datetime.date(year_, month_, day_)
    for i in range(range_of_days):
        date = start + datetime.timedelta(days=i)
        params = {
            "date": date.strftime("%Y%m%d"),
            "json": "json"
        }
        r = requests.get(
            'https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange',
            params=params).json()
        for x in r:
            if x['cc'] == 'USD':
                x['rate'] = float(x['rate'])
                list_month_rate.append(x['rate'])
    return list_month_rate


def mean(sample: float) -> float:
    """Calculate the mean value of sample data.
    : param `sample`: the investigated data
    """
    mean = round(sum(sample)/len(sample), 4)
    return mean
    

def st_deviation(sample: float) -> float:
    """Calculate the standard deviation of sample data.
    :param `sample`: the investigated data
    """
    n = len(sample)
    mean = sum(sample) / n
    variance = sum([((num - mean) ** 2) for num in sample]) / n
    std_dev = round(math.sqrt(variance), 4)
    return std_dev


dict_USD_rate = {}
    
list_March_2022 = gathering_info(2022, 3, 1, 31)
dict_USD_rate['Березень 2022'] = {
    'середній курс USD': mean(list_March_2022),
    'відхилення курсу USD': st_deviation(list_March_2022)
    }

list_April_2022 = gathering_info(2022, 4, 1, 30)
dict_USD_rate['Квітень 2022'] = {
    'середній курс USD': mean(list_April_2022),
    'відхилення курсу USD': st_deviation(list_April_2022)
    }

list_May_2022 = gathering_info(2022, 5, 1, 31)
dict_USD_rate['Травень 2022'] = {
    'середній курс USD': mean(list_May_2022),
    'відхилення курсу USD': st_deviation(list_May_2022)
    }

list_June_2022 =  gathering_info(2022, 6, 1, 30)
dict_USD_rate['Червень 2022'] = {
    'середній курс USD': mean(list_June_2022),
    'відхилення курсу USD': st_deviation(list_June_2022)
    }

list_July_2022 = gathering_info(2022, 7, 1, 31)
dict_USD_rate['Липень 2022'] = {
    'середній курс USD': mean(list_July_2022),
    'відхилення курсу USD': st_deviation(list_July_2022)
    }

list_August_2022 = gathering_info(2022, 8, 1, 31)
dict_USD_rate['Серпень 2022'] = {
    'середній курс USD': mean(list_August_2022),
    'відхилення курсу USD': st_deviation(list_August_2022)
    }

list_September_2022 = gathering_info(2022, 9, 1, 30)
dict_USD_rate['Вересень 2022'] = {
    'середній курс USD': mean(list_September_2022),
    'відхилення курсу USD': st_deviation(list_September_2022)
    }

list_October_2022 = gathering_info(2022, 10, 1, 31)
dict_USD_rate['Жовтень 2022'] = {
    'середній курс USD': mean(list_October_2022),
    'відхилення курсу USD': st_deviation(list_October_2022)
    }

list_November_2022 = gathering_info(2022, 11, 1, 30)
dict_USD_rate['Листопад 2022'] = {
    'середній курс USD': mean(list_November_2022),
    'відхилення курсу USD': st_deviation(list_November_2022)
    }

list_December_2022 = gathering_info(2022, 12, 1, 31)
dict_USD_rate['Грудень 2022'] = {
    'середній курс USD': mean(list_December_2022),
    'відхилення курсу USD': st_deviation(list_December_2022)
    }

list_January_2023 = gathering_info(2023, 1, 1, 31)
dict_USD_rate['Січень 2023'] = {
    'середній курс USD': mean(list_January_2023),
    'відхилення курсу USD': st_deviation(list_January_2023)
    }

list_February_2023 = gathering_info(2023, 2, 1, 28)
dict_USD_rate['Лютий 2023'] = {
    'середній курс USD': mean(list_February_2023),
    'відхилення курсу USD': st_deviation(list_February_2023)
    }

print(dict_USD_rate)
```

### Завдання 2:

Потрібно проаналізувати всі товари на сайті: 
https://smallpacking.agrosem.ua/products

Створіть базу даних, що містить таблицю із такими стовбцями: 
    - артикул товару
    - найменування
    - ціна 
    - вага

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import json

def info_products(python_dicts):  
    product_dict = {}
    id_= python_dicts["id"]
    n = python_dicts['name']
    p = python_dicts['price']
    w = python_dicts['weight']
    product_dict={
        "id_": id_,
        "name": n,
        "price": p,
        "weight": w
    }
    return  product_dict


driver_path = 'C:/Users/liman/Desktop/chromedriver_win32/chromedriver.exe'
options = webdriver.ChromeOptions()
options.add_argument('--headless')  
driver = webdriver.Chrome(driver_path, options=options)

for page in range(1, 10):
    url = "https://smallpacking.agrosem.ua/products/?page={page}".format(page = page)
    driver.get(url)
    driver.implicitly_wait(10)
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    for y in soup.find_all("div", class_='item'):
        json_dicts = y.get('content')
        python_dicts = json.loads(json_dicts)
        print(info_products(python_dicts))
    
driver.quit()
```
