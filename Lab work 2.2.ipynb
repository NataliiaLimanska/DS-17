### Завдання 6 

Напишіть функції, що:
* розраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел генеруємо рандомно

```python
import random

#генеруємо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]


def find_min_max(numbers):
    if not numbers:
        return None, None
    else:
        min_num = min(numbers)
        max_num = max(numbers)
        return (min_num, max_num)

print(f'Мінімальне та максимальне число: {find_min_max(list_with_random_numbers)}')
```

```python
def find_mean(numbers):
    mean = sum(numbers) / len(numbers)
    return mean

print(f'Середнє значення послідовності: {find_mean(list_with_random_numbers)}')
```

```python
def find_median(numbers):
    n = len(numbers)
    sorted_numbers = sorted(numbers)
    if n % 2 == 0:
        median = (sorted_numbers[n//2-1] + sorted_numbers[n//2]) / 2
    else:
        median = sorted_numbers[n//2]
    return median

print(f'Медіана послідовності: {find_median(list_with_random_numbers)}')
```

```python
def find_frequencies(numbers):
    dict_of_nums = {}
    for num in numbers:
        if num in dict_of_nums:
            dict_of_nums[num] +=1
        else:
            dict_of_nums[num] = 1
    return dict_of_nums
    
print(f'Частота кожної цифри у послідовності: {find_frequencies(list_with_random_numbers)}')
```

```python
import math


def st_deviation(numbers):
    n = len(numbers)
    mean = sum(numbers) / n
    variance = sum([((num - mean) ** 2) for num in numbers]) / n
    std_dev = math.sqrt(variance)
    return std_dev

print(f'Середнє квадратичне відхилення: {st_deviation(list_with_random_numbers)}')
```
